```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
#| label: detection-zones
#| fig-cap: ""
#| fig-align: "center"
#| echo: false

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(5, 8), subplot_kw={'projection': 'polar'})

# Dibujar las divisiones de las zonas
ax.vlines(np.deg2rad(np.linspace(0, 180, 6)), 0, 15, colors='C3', linestyles='solid')

# Dibujar las etiquetas para cada zona
for name, angle in zip(["R", "FR", "F", "FL", "L"], np.deg2rad(np.linspace(0, 180, 11)[1::2])):
    plt.text(angle, 10, name, size=10, rotation=0,
            ha="center", va="center",
            bbox=dict(boxstyle="round",
                    ec=(1., 0.5, 0.5),
                    fc=(1., 0.8, 0.8),
                    )
            )

# Rango radial
ax.set_rmax(15.0)
ax.set_rmin(0.05)
ax.set_rorigin(0)  # Set the origin of the radial axis to 0

# Rango angular
ax.set_thetamin(0)
ax.set_thetamax(180)

# Eliminar labels del eje radial
ax.set_yticklabels([])

# Cuadr챠culas
ax.minorticks_on()
# ax.grid(True, which='major', linestyle='--', linewidth=0.5)
# ax.grid(True, which='minor', linestyle=':', linewidth=0.3)
ax.grid(True)

# T챠tulo y leyenda
plt.title('Zonas de detecci처n de obst찼culos')
# ax.legend(loc='upper left')

plt.tight_layout()
plt.show()
```
